### 명사만으로 구성된 새로운 문장을 생성

## 단어 임베딩
### Word2Vec()
    - CBOW 방식과 Skip-Gram 방식의 단어 임베딩을 구현한 C++ 라이브러리
    - 파이썬에서는 gensim이라는 패키지에 Word2Vec이라는 클래스로 구현
    - 기본적인 임베딩 모형에 subsampling, negative sampling 등의 기법을 추가하여 학습 속도를 향상
- 인자 size : 단어(특징) 벡터의 수 
    - 보통 50~300
    - 단어 벡터의 크기를 크게 잡으면 그만큼 정교하게 단어를 표현할 수 있지만 
    - 벡터의 크기가 클수록 학습하는데 많은 문장이 필요
- most_similar() : 가장 유사한 단어를 출력
- similarity() : 두 단어의 유사도 계산

```python
# 일반 PC, 2~3분 소요

model = Word2Vec(sentence_list, size=100)   # size = 특징 벡터의 수
```

```python
model.most_similar("한국")
```

```python
print(model.similarity("트럼프","미국"))
print(model.similarity("한국","미국"))
```

### 단어의 유사도
- #### 두 개의 문자열이 얼마나 다른지를 나타내는 편집 거리를 이용 
- #### 편집 거리 
    - 한 단어에서 글자를 추가, 제거, 변경함으로써 다른 단어로 바꿀 때 필요한 최소한의 편집 행동의 횟수로 결정
    - NLTK 라이브러리를 활용하여 두 문장의 편집거리 계산이 가능하
        - nltk.edit_distance(i, j)

```python
import nltk

sentence_list = ["우리 모두 함께 놀자", "모두 같이 놀자", "놀자", "모두 다 같이"]

for i in sentence_list:
    print("'", i, "'")
    
    for j in sentence_list:
        print("\t", j, " : ", end='')
        print(nltk.edit_distance(i, j))
        
    print()
```

### 코사인 유사도- 직접 구현

```python
# 직접 구현시 코사인 유사도
import math
def cosine_similarity(v1,v2):
    
    sumxx, sumxy, sumyy = 0, 0, 0
    for i in range(len(v1)):
        x = v1[i]; y = v2[i]
        sumxx += x*x
        sumyy += y*y
        sumxy += x*y
    return sumxy/math.sqrt(sumxx*sumyy)
```

### 자카드 유사도- 직접 구현
- #### str1.split()
- #### a.intersection(b)

```python
# 자카드 유사도 계산

sentence_list = ['우리 모두 함께 놀자','모두 함께 놀자']

def get_jaccard_sim(str1, str2): 
    a = set(str1.split()) 
    b = set(str2.split())
    c = a.intersection(b)
    return float(len(c)) / (len(a) + len(b) - len(c))
```

```python
get_jaccard_sim(sentence_list[0], sentence_list[1])
```

```python
from nltk import ngrams

sentence = 'this is a foo bar sentences and i want to ngramize it'

n = 2
sixgrams = ngrams(sentence.split(), n)
for grams in sixgrams:
    print(grams)
```
