### 문서-단어 행렬 만들기

- 전체 문서 {d1,d2,…,dn}를 구성하는 고정된 단어장(vocabulary) {t1,t2,…,tm}를 만든다.
- 각 di 개별 문서에 단어장에 해당하는 단어들이 포함되어 있는지를 표시한다.
- x(i,j) = 문서 di 내의 단어 tj의 출현 빈도
- 또는
- x(i,j) = {0,1}
    - 단어 tj가 문서 di 안에 없으면 : 0
    - 단어 tj가 문서 di 안에 있으면 : 1

### CountVectorizer()
- 1. 문서를 토큰(단어) 리스트로 변환한다.
- 2. 각 문서에서 토큰(단어)의 출현 빈도를 센다. : vocabulary_
- 3. 각 문서를 BOW 인코딩 벡터로 변환한다.--> {0, 1}

#### (인수 설명)
- stop_words : 문자열 {‘english’}, 리스트 또는 None (디폴트)
    - stop words 목록.‘english’이면 영어용 스탑 워드 사용
- ngram_range : (min_n, max_n) 튜플
    - n-그램 범위
- max_df : 정수 또는 [0.0, 1.0] 사이의 실수. 디폴트 1
    - 단어장에 포함되기 위한 최대 빈도
- min_df : 정수 또는 [0.0, 1.0] 사이의 실수. 디폴트 1
    - 단어장에 포함되기 위한 최소 빈도

### TF-IDF(Term Frequency – Inverse Document Frequency) 인코딩
- 단어를 갯수 그대로 카운트하지 않고 모든 문서에 공통적으로 들어있는 단어의 경우 문서 구별 능력이 떨어진다고 보아 가중치를 축소하는 방법
- 구제적으로는 문서 d(document)와 단어 t 에 대해 다음과 같이 계산
- tf-idf(d,t)=tf(d,t)⋅idf(t)
    - tf(d,t) : term frequency. 
        - 특정한 단어의 빈도수
    - idf(t)  : inverse document frequency. 
        - 특정한 단어가 들어 있는 문서의 수에 반비례하는 수

- idf(d,t)=log(n/(1+df(t)))
- n  : 전체 문서의 수
- df(t) : 단어 t 를 가진 문서의 수

```python
tv = TfidfVectorizer(max_features=2000)
dtm = tv.fit_transform(news_text.tolist())
df = pd.DataFrame(dtm.toarray(), columns = tv.get_feature_names())
df
```
